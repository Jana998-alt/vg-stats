# Video Game Sales

from _https://www.kaggle.com/gregorut/videogamesales_ 

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
vg_df = pd.read_csv('vgsales.csv')
```

```python
vg_df
```

```python
# Q1: Which company is the most common video game publisher?
vg_df['Publisher'].value_counts().idxmax()
```

```python
#  Q2: What’s the most common platform?
vg_df['Platform'].value_counts().idxmax()
```

```python
#  Q3: What is the most common genre?
vg_df['Genre'].value_counts().idxmax()
```

```python
#  Q4: What are the top 20 highest grossing games?
vg_df.nlargest(20, 'Global_Sales')
```

```python
# Q5: For North American video game sales, what’s the median?
# Provide a secondary output showing ten games surrounding the median sales output
# assume that games with same median value are sorted in descending order

vg_df['NA_Sales'].median()
x = vg_df[vg_df['NA_Sales']==vg_df['NA_Sales'].quantile(interpolation='nearest')]
x.iloc[225:235]
```

```python
# Q6: For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?
# Z-score = (value - mean)/std
(vg_df['Global_Sales'].max() - vg_df['NA_Sales'].mean())/ vg_df['NA_Sales'].std()
```

```python
# Q7:  The Nintendo Wii seems to have outdone itself with games,
# How does its average number of sales compare with all of the other platforms?

mean_Wii_sales = vg_df.loc[vg_df['Platform'] == 'Wii']['Global_Sales'].mean()
```

```python
mean_Wii_sales
```

```python
comparing_Wii_with_other_platforms_df = vg_df.groupby('Platform', as_index=False)['Global_Sales'].mean()
```

```python
comparing_Wii_with_other_platforms_df['Compare_with_Wii'] =mean_Wii_sales/comparing_Wii_with_other_platforms_df['Global_Sales'] 
```
